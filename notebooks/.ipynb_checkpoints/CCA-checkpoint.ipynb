```python
from sklearn.cross_decomposition import CCA
X = [[0., 0., 1.], [1.,0.,0.], [2.,2.,2.], [3.,5.,4.]]
y = [[0.1, -0.2], [0.9, 1.1], [6.2, 5.9], [11.9, 12.3]]
cca = CCA(n_components=1)
cca.fit(X, y)
X_c, Y_c = cca.transform(X, y)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
X = [[0., 0., 1.], [1., 0., 0.], [2., 2., 2.], [3., 5., 4.]]
y = [[0.1, -0.2], [0.9, 1.1], [6.2, 5.9], [11.9, 12.3]]

# Initialize and fit the CCA model
cca = CCA(n_components=1)
cca.fit(X, y)

# Transform the data using the fitted CCA model
X_c, Y_c = cca.transform(X, y)

# Plot the transformed data
plt.figure(figsize=(8, 6))

# Plot the transformed X
plt.subplot(1, 2, 1)
plt.scatter(X_c, [0] * len(X_c), c='b', label='Transformed X')
plt.xlabel('Canonical Component 1')
plt.ylabel('X')
plt.title('Transformed X')
plt.legend()

# Plot the transformed y
plt.subplot(1, 2, 2)
plt.scatter(Y_c, [0] * len(Y_c), c='r', label='Transformed y')
plt.xlabel('Canonical Component 1')
plt.ylabel('y')
plt.title('Transformed y')
plt.legend()

plt.tight_layout()
plt.show()
```
